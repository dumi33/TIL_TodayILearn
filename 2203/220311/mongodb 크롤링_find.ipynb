### find의 다양한 문법 - nor
- not or

```python
docs = actor_collection.find({'$nor' : [{'흥행지수' : {'$gte' : 10000}},{'흥행지수' : {'$lte' : 2000}}]},{'배우이름':1,'흥행지수' : 1,'_id' : 0})
for doc in docs :
    print(doc)
```

### find의 다양한 문법 - in, nin
- in : 들어가있다.
- nin : not in : 들어가 있지 않다.

```python
docs = actor_collection.find({'흥행지수' : {'$in' : [8329,8956]}},{'배우이름':1,'흥행지수' : 1,'_id' : 0})
for doc in docs :
    print(doc)
```

### find의 다양한 문법 - elemMatch
* 적어도 한 개 이상의 리스트 요소가 복수 개의 조건을 동시에 만족하는 경우

```python
from bs4 import BeautifulSoup
import requests
import pymongo
import re # 정규표현식을 위해 

conn = pymongo.MongoClient()
actor_db = conn.cine21
elemmatch_sample = actor_db.sample
```

```python
elemmatch_sample.insert_many([
    {'results' : [82,85,88]},
    {'results' : [75,88,91]}
])
```

```python
docs = elemmatch_sample.find({'results': {'$gte' : 90, '$lt' : 80}})
for doc in docs :
    print(doc)
```

```python
docs = elemmatch_sample.find({'results': {'$elemMatch' : {'$gte' : 75,'$lt' : 80}}})
for doc in docs :
    print(doc)
```
